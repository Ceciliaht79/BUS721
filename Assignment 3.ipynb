```python
import numpy as py
import pandas
import statsmodels.api as sm
from statsmodels.tsa.ar_model import AutoReg
import matplotlib.pyplot as plot
```

```python
noise = np.random.randn(1000)
white_noise = (noise-np.mean(noise))/np.std(noise)
```

```python
Y=[0] #y0=0

for i in range(25):
    yt = 1 +0.8*Y[i]+white_noise[i]
    Y.append(yt)
    
Y_lag=[1]

for i1 in range(25):
    lag_yt = 1 +0.8*Y_lag[i1]+white_noise[i1]
    Y_lag.append(lag_yt)

AR_model = AutoReg(Y,1).fit()
predict = AR_model.predict(2,25)
residual = Y[2:] - predict
plot.plot(residual)
```

```python
Y_lag = sm.add_constant(Y_lag)
model_ols = sm.OLS(Y,Y_lag).fit()
print(model_ols.summary())
```
